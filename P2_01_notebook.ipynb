On parlera ensuite de 'Valeurs atypiques' les valeurs comprisent entre le 3e et le 4e quartille et n'étant pas des valeurs aberrantes. En anglais ces deux définissions sont comprises dans le terme 'outlier'.

> # II Traitement des données

Nous avons pu identifier les colonnes essentielles à l'étude des arbres, nous avons également repéré les valeurs manquantes, aberrantes et atypiques. À présent, nous allons nettoyer le jeu de données avant de pouvoir passer à son étude approfondie.

### Suppression des colonnes inutiles

Nous avons identifié précédement les colonnes que nous souhaite supprimer, les colonnes : "numéro" qui est vide,
la colonne "libellé_français", qui contient des valeurs manquantes dons des traductions manquantes, les colonnes "id" et "id_emplacement"
et enfin la colonne 'type_emplacement'.

```python
#Suppression des colonnes non-pertinentes
df = df.drop(['numero','libelle_francais','id_emplacement','id','type_emplacement' ],axis=1)
```

### Suppression des valeurs dupliquées

Comme vus précédemment nous avons défini qu'une ligne était dupliquée si elle possédait des valeurs de coordonnées, hauteur et circonférence identiques.

```python
#Suprression des lignes dupliquées
df = df.drop(df[df.geo_point_2d_a.duplicated() & df.geo_point_2d_b.duplicated() & df.espece.duplicated() & 
                df.hauteur_m.duplicated()].index, axis=0)
```

### Remplacement des valeurs nulles

Afin de coller au mieux à la réalité des valeurs relevées dans ce jeu de données nous avons choisi de remplacer chaque valeur nulle dans la colonnes "hauteur_m" et "circonférence_cm" par les moyennes de hauteur et circonférence des espèces auxquelles ces arbres appartiennent. 

```python
#Désactivation des messages d'alerte
warnings.filterwarnings('ignore')

# Remplacement des valeurs nulles pas la hauteur moyenne de l'espèce d'apartenance
for u in df.genre.unique():
    df.hauteur_m.loc[(df.genre == u) & (df.hauteur_m == 0)] = np.mean(df[df.genre == u].hauteur_m)

# Remplacement des valeurs nulles pas la circonference moyenne de l'espèce d'apartenance
for u in df.genre.unique():
    df.circonference_cm.loc[(df.genre == u) & (df.circonference_cm == 0)] = np.mean(df[df.genre == u].circonference_cm)
    

# Ici on modifie l'expression des valeurs dans la colonne 'arrondissement' pour améliorer le rendu dans les graphiques
df.arrondissement = df.arrondissement.apply(lambda x : x.split('PARIS ')[-1])
```

### Définition de l'intervalle interquartile

> #### Approche Méthodologique
Nous souhaitons définir notre plage de sélection pour étudier nos données lors de notre approche approfondie. Pour cela, nous devons sélectionner les valeurs les plus fiables possibles.
Grâce à l'analyse des hauteurs et circonférence et aux quartiles de distribution, nous pouvons poser un critère d'acceptation définie par l'écart-interquartile (EI):
<br>EI = Q3 - Q1
<br>Hauteur <= Q3(hauteur) + 1.5 * EI(hauteur)
<br>Circonférence <= Q3(circonf) + 1.5 * EI(circonf)

```python
##Hauteur
# Calcul de l'écart incterquantile de la hauteur
EI_hauteur = hauteur.quantile(.75) - hauteur.quantile(.25)

#Définition du premier critère de hauteur
C_hauteur = hauteur.quantile(.75) + 1.5 * EI_hauteur

##Crirconférence
# Calcul de l'écart incterquantile de la circonférence
EI_circonference = circonf.quantile(.75) - circonf.quantile(.25)
#Définition du second critère de circonférence
C_circonf = circonf.quantile(.75) + 1.5 * EI_circonference


print(f'Nous sélectionnons les arbres ayant une hauteur maximal de {C_hauteur}m et une circonférence maximale de {C_circonf}cm.')
```

```python
# Sélection de la plage de données d'étude appronfoondie en fonction des critères définis précédement
df = df[(df.hauteur_m <= C_hauteur) & (df.circonference_cm <= C_circonf)]


print(f"Le nétoyage de la base de données a necéssiter la suppression de {200137 - len(df)} lignes soit moins de 4.05% des données")
```

Dans cette partie nous reviendrons plus en détail sur les différentes informations délivrées par le jeu de donnée. ### Analyse du stade de développement

L'objectif dans cette partie est d'analyser la réparation des arbres en prenant comme critère principal le stade de développement.

```python
# On crée ici pour plus de visibilité un dictionnaire contenant en clée le stade de développement et en valeur sa signification
stade_developpement = {
    'A' : 'Adulte',
    'JA' : 'Jeune Adulte',
    'J' : 'Jeune',
    'M' : 'Mature'

}

# On affiche dans un graphique 'camembert' la répartion du stade de développement
plt.figure(figsize=(7,7))
plt.title('figure 5 : Répartition du stade de développement', fontsize=15, fontweight='bold')
plt.pie(df.stade_developpement.value_counts(),labels=stade_developpement.values(),autopct='%1.2f%%')
plt.legend()
plt.show()
```

On observe que près de la moitié des arbres de Parisien sont des arbres adultes, les jeunes arbres représentent quant à eux 20% du parc implanté et près de 3% des arbres sont matures.

```python
#On cherche ici à afficher la répartition du stade de développement en fonction de la hauteur des arbres
plt.figure(figsize=(15, 8))
plt.title("figure 6 : Nombre d'arbre en fonction de la hauteur et du stade de développement", fontsize=15, fontweight='bold')
for u in stade_developpement.keys():
    plt.hist(df[df.stade_developpement == u].hauteur_m,range = (0, 30),bins=35,label=stade_developpement[u])
plt.legend()
plt.xlabel('Hauteur (m)', fontsize=14)
plt.ylabel("Nombre d'arbre", fontsize=14)
plt.show()
```

Majoritairement, on constate que les arbres les moins hauts, inférieurs à 7 mètres, sont des arbres jeunes, et inversement les arbres les plus grands, supérieurs à 15 mètres, sont des arbres adultes ou matures.

```python
#On affiche le ombre d'arbre en fonction de la hauteur et du stade de développement
plt.figure(figsize=(15, 8))
plt.title("figure 7 : Répartition des arbres en fonction de la hauteur et du stade de développement", fontsize=15, fontweight='bold')
for u in stade_developpement.keys():
    plt.hist(df[df.stade_developpement == u].circonference_cm,range = (0, 300),bins=35,label=stade_developpement[u])

plt.xlabel('Circonférence (cm)', fontsize=14)
plt.ylabel("Nombre d'arbre", fontsize=14)
plt.legend()
plt.show()
```

L'observation est encore plus significative avec la répartition de la circonférence des arbres segmentés en fonction de leur stade d'évolution. On remarque une corrélation entre la circonférence de l'arbre et son stade de développement : plus l'arbre est jeune plus sa circonférence est petite, plus l'arbre a un stade de développement avancé plus sa circonférence sera élevée.

```python
#On affiche ici la Hauteur des arbres en fonction de leur circonférence et du stade de développement
plt.figure(figsize=(15, 8))
plt.title('figure 8 : Hauteur des arbres en fonction de leur circonférence et du stade de développement', 
          fontsize=15,
          fontweight='bold')
sns.scatterplot(y='hauteur_m',
                x='circonference_cm',
                data=df,
                hue="stade_developpement")
plt.legend()
plt.show()
```

Sur ce graphique, on constate la relation entre la hauteur, la circonférence et le stade de développement :
Plus l'arbre est jeune plus sa taille et sa hauteur seront faible.

Ce graphique est un bon outil pour visualiser l'état de développement des arbres, mais également pour identifier d'éventuelles valeurs aberrantes, en effet, on distingue clairement des populations distinctes en fonction du stade de développement.

### Analyse en fonction des espèces

L'objectif de cette partie est d'observer la population d'arbre parisien en fonction de leur espèce.

Pour rappel, on compte 175 genres d'arbre différents d'après le tableau 2.
